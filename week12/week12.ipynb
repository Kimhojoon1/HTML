```python
import csv

file = open("./seoul.csv", "r", encoding='cp949')
data = csv.reader(file)
Data = next(data)
lowest = float(Data[3])
highest = float(Data[4])

Range = highest - lowest
highDate = Data[0]
lowDate = Data[0]
bigRange = Data[0]

for k in data:
    tempRange = float(k[4])-float(k[3])
    if highest < float(k[4]):
        highest = float(k[4])
        highDate = k[0]
    if lowest > float(k[3]):
        lowest = float(k[3])
        lowDate = k[0]
    if Range < tempRange:
        Range = tempRange
        bigRange = k[0]

file.close()

print("일교차가 가장 큰 날")
print(bigRange)
print("최저기온이 가장 낮은 날")
print(lowDate)
print("최고기온이 가장 높은 날")
print(highDate)
```
